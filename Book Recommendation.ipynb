```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
```

```python
books = pd.read_csv("books.csv")
users = pd.read_csv("users.csv")
ratings = pd.read_csv("ratings.csv")
```

## Popularity based reccomednation

```python
ratings_with_name=ratings.merge(books,on="ISBN",how="inner")
```

```python
ratings_with_name.head()
```

```python
num_rating_df=ratings_with_name.groupby("Book-Title")["Book-Rating"].count().reset_index()
num_rating_df.rename(columns={"Book-Rating":"num_rating"},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popularity_df
```

```python
popularity_df=popularity_df[popularity_df["num_rating"]>=250].sort_values("avg_rating",ascending=False).head(50)
```

```python
popularity_df
```

```python
popularity_df = popularity_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Year-Of-Publication",
                                                                          "num_rating","avg_rating","Image-URL-M"]]
```

```python
popularity_df.head()
```

```python
import pickle
pickle.dump(popularity_df,open("popular.pkl","wb"))
```

## Collabarative Filtering  

```python
 x = ratings_with_name.groupby("User-ID")["Book-Rating"].count() >200
top_users = x[x].index #boolean indexing
```

```python
filtertd_rating = ratings_with_name[ratings_with_name["User-ID"].isin(top_users)]
```

```python
filtertd_rating.head()
```

```python
 y = filtertd_rating.groupby("Book-Title")["Book-Rating"].count() >= 50
famous_books = y[y].index
```

```python
final_ratings = filtertd_rating[filtertd_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings = final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
smliarity_score = cosine_similarity(pt)
```

```python
smliarity_score
```

```python
books.head()
```

```python
def reccomede(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similiar_items = sorted(list(enumerate(smliarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similiar_items:
        item=[]
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Year-Of-Publication"].values))

        data.append(item)

    return data

    

```
